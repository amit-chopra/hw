```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data

city_df = pd.read_csv(city_data_to_load)
city_df.head()
```

```python
ride_df = pd.read_csv(ride_data_to_load)
ride_df.head()
# Combine the data into a single dataset

merge_pd = pd.merge(ride_df, city_df, on='city')

# Display the data table for preview
merge_pd
```

```python
count_lake = 0
counts = {city: 0 for city in merge_pd["city"]}
for index, row in merge_pd.iterrows():
    counts[merge_pd.loc[index, "city"]] += 1
    if merge_pd.loc[index, "city"] == "Lake Jonathanshire":
        count_lake+=1
print(count_lake)
print(counts)
testlist = []
for key, value in counts.items():
    testlist.append((key,value))
testlist
```

```python
count_df = pd.DataFrame(testlist, columns=["City", )
count_df
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
city_grp_by = merge_pd.groupby(['type','city'])
city_grp_by.sum()
```

```python
#Urban Ride Per City
total_no_rides =city_grp_by['ride_id'].count()
urban_rides_per_city = merge_pd[merge_pd["type"] == "Urban"].groupby("city")["ride_id"].count()
#total_no_rides
urban_rides_per_city
```

```python
# Rural Ride Per City
total_no_rides =city_grp_by['ride_id'].count()
rural_rides_per_city = merge_pd[merge_pd["type"] == "Rural"].groupby("city")["ride_id"].count()
#total_no_rides
rural_rides_per_city
```

```python
# Subu Ride Per City
total_no_rides =city_grp_by['ride_id'].count()
subu_rides_per_city = merge_pd[merge_pd["type"] == "Suburban"].groupby("city")["ride_id"].count()
#total_no_rides
subu_rides_per_city
```

```python
# Urban Fare
urban_fare = merge_pd[merge_pd["type"] == "Urban"].groupby("city")["fare"].mean()
urban_fare
```

```python
# Rural Fare
rural_fare = merge_pd[merge_pd["type"] == "Rural"].groupby("city")["fare"].mean()
rural_fare
```

```python
# Subu Fare
subu_fare = merge_pd[merge_pd["type"] == "Suburban"].groupby("city")["fare"].mean()
subu_fare
```

```python
# Build the scatter plots for each city types
Urban = plt.scatter(urban_rides_per_city, urban_fare,marker="o",s=20*urban_rides_per_city,facecolors="pink", edgecolors="black", alpha=0.7, label='Urban')

# Build the scatter plots for each city types
Sub_U =plt.scatter(subu_rides_per_city, subu_fare,marker="o",s=10*subu_rides_per_city,facecolors="yellow", edgecolors="black", alpha=0.7, label= 'Sub_U')   
    
# Build the scatter plots for each city types
Rural = plt.scatter(rural_rides_per_city, rural_fare,marker="o",s=10*rural_rides_per_city,facecolors="blue",edgecolors="black",alpha=0.25, label= 'Rural')

# Incorporate the other graph properties
plt.title('Pyber Ride Sharing Data 2016')
plt.xlabel('Total Number of Rides (per City)')
plt.ylabel('Average Fare($)')
plt.grid()
plt.text(45, 30, "Note:color size correlates to driver count per city")

# Create a legend
plt.legend(title= 'City Types')

# Incorporate a text label regarding circle size
plt.savefig('../images/pyber_ride_share')
# Save Figure
```

## Total Fares by City Type

```python
# Calculate Type Percents
urb = merge_pd[merge_pd['type']=='Urban']['fare'].sum()
urb
```

```python
rur = merge_pd[merge_pd['type']=='Rural']['fare'].sum()
rur
```

```python
subu = round(merge_pd[merge_pd['type']=='Suburban']['fare'].sum(), 2)
subu
```

```python
rides = [urb,rur,subu]
labels = ['Urban', 'Rural', 'Suburban']
colors = ["yellowgreen","lightskyblue","lightcoral"]
explode = (0.1,0,0)
```

```python
# Build the pie chart
plt.pie(rides, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=False, startangle=300)
plt.title('% of Total Fares by City Type')
plt.axis("equal")
# Save Figure
plt.savefig('../images/pieinsky.png')
# Show Figure
plt.show()
```
